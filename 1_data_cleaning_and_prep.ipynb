## DATA CLEANING

Firstly Removed Unnecessary columns then, Data was cleaned by converting columns to numeric format and handling missing values using median and group-wise median techniques.

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
```

```python
import re
def extract_no(x):
    x=str(x)
    match=re.search(r"\d+\.?\d*",x)
    if match:
        return float(match.group())
    return None

df["mileage"]=df["mileage"].apply(extract_no)
df["engine"]=df["engine"].apply(extract_no)
df["max_power"]=df["max_power"].apply(extract_no)
```

```python
import re
def clean_torque(x):
    x=str(x).lower().strip()
    match=re.search(r"\d+\.?\d*",x)
    if match:
        return float(match.group())
    return None
    number=float(match.group())
    if "kgm" in x:
        return round(number * 9.80665,2)
    return number

df['torque']=df['torque'].apply(clean_torque)
```

### UNDERSTANDING DATA DISTRIBUTION

Before filling missing values,I analyzed the data distribution.
Most values lie in a normal range.
Some extreme values(outliers) are present.
These outliers can distort the mean.

### CHOOSING MEDIAN OVER MEAN

Instead of mean,I used median for imputation because:
Median is more robust and stable.
It represents central tendency better in skewed data.

### APPROACH USED

For Mileage,Engine,Max Power used group-wise median(based on Fuel Type)
For Torque & Seats,used overall median(less variation)

```python
df['mileage']=df.groupby('fuel')['mileage'].transform(lambda x:x.fillna(x.median()))
df['engine']=df.groupby('fuel')['engine'].transform(lambda x:x.fillna(x.median()))
df['max_power']=df.groupby('fuel')['max_power'].transform(lambda x:x.fillna(x.median()))
df['torque'].fillna(df['torque'].median(),inplace=True)
df['seats'].fillna(df['seats'].median(),inplace=True)
```

##  Owner Data Cleaning

Owner categories were simplified by grouping higher ownership levels into a single category to reduce sparsity and improve analysis.

```python
df['owner'].value_counts()
```

```python
def clean_owner(x):
    if x == "Test Drive Car":
        return "First Owner"
    elif x == "Fifth":
        return "Fourth & Above Owner"
    else:
        return x

df['owner']=df['owner'].apply(clean_owner)
```

### Outlier Detection (IQR Method)

```python
Q1=df['selling_price'].quantile(0.25)
Q3=df['selling_price'].quantile(0.75)
IQR=Q3-Q1

upper_fence=Q3+1.5*IQR
lower_fence=Q1-1.5*IQR

df['is_outlier']=(df['selling_price']>upper_fence)| (df['selling_price']<lower_fence)
```

"I used the Interquartile Range (IQR) method to identify extreme price values in the selling_price column. Instead of removing this data, I added an is_outlier flag to ensure that luxury car prices do not skew the average market trends during my EDA and SQL analysis."
